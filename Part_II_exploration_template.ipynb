# Dataset Overview and Executive Summary
The Titanic dataset contains information for 891 passengers and 15 variables.

The dataset includes demographic information, passenger class, fare, embarkation details, and survival status.

This presentation highlights the most important findings discovered during exploratory data analysis using a small number of polished visualizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style("whitegrid")

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("dataset.csv")

df.head()
```

# Visualization 1
## Passenger Age Distribution
This visualization shows the overall age distribution of Titanic passengers.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="age",
    bins=20,
    color="steelblue"
)

plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

## Observation
Most passengers were between 20 and 40 years old. The distribution is slightly right-skewed, with fewer elderly passengers.

# Visualization 2
## Fare Distribution by Passenger Class
This visualization compares the fare paid by passengers in each travel class.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="class",
    y="fare"
)

plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.show()
```


## Observation
Passengers traveling in First Class generally paid significantly higher fares than passengers in Second and Third Class.

# Visualization 3
## Age vs Fare by Survival Status
This visualization explores how age and fare relate to passenger survival.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="age",
    y="fare",
    hue="survived",
    alpha=0.7
)

plt.title("Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

### Observation
Passengers who paid higher fares were more likely to survive. Age alone does not show a strong relationship with survival.
