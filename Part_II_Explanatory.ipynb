```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
ordinal_var_dict = {
    'cut':['Fair','Good','Very Good','Premium','Ideal'],
    'color':['J','I','H','G','F','E','D'],
    'clarity':['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
}

for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(
        categories=ordinal_var_dict[var],
        ordered=True
    )
    diamonds[var] = diamonds[var].astype(ordered_var)
```

```python
bins = 10**np.arange(2.4,4.4+0.05,0.05)

plt.figure(figsize=(8,5))

plt.hist(diamonds['price'],bins=bins)

plt.xscale('log')

plt.xticks([500,1000,2000,5000,10000,20000],
           ['500','1000','2000','5000','10000','20000'])

plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.title("Distribution of Diamond Prices")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(diamonds['carat'],bins=40)

plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.title("Distribution of Carat")

plt.show()
```

```python
sample = diamonds.sample(3000, random_state=42)

plt.figure(figsize=(8,6))

sns.scatterplot(data=sample,
                x='carat',
                y='price',
                alpha=0.4)

plt.title("Price vs Carat")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=diamonds,
            x='cut',
            y='price')

plt.title("Diamond Price by Cut")

plt.show()
```

```python
sample = diamonds.sample(3000, random_state=42)

plt.figure(figsize=(10,6))

sns.pointplot(data=sample,
              x='clarity',
              y='price',
              hue='color',
              errorbar=None)

plt.title("Price by Color and Clarity")

plt.show()
```
